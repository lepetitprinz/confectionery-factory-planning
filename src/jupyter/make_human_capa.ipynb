```python
import numpy as np
import pandas as pd
```

### People Capacity

```python
res_org = ['10003935','10003936','10003977','10003978','10003980','10003981','10007202']
res_cd = []
for res in res_org:
    for number in range(1,11):
        res_cd.append(res + '_M' + str(number).zfill(3))
        res_cd.append(res + '_F' + str(number).zfill(3))
res_cd = sorted(res_cd)
```

```python
project_cd = 'ENT_001'
if_vrsn_id = '-'
plant_cd = 'K120'
res_seq = 4
res_type_cd = 'PPL'
res_ctgi_cd = ''
res_grp_cd = ['2217'] * len(res_cd)
res_capa_val = [1] * len(res_cd)
over_capa_val = [0] * len(res_cd)
capa_unit_cd = 'MIN'
start_yymmdd = '20010101'
end_yymmdd = '99991231'
descr = ''
inf_capa_yn = 'N'
use_yn = 'Y'
del_yn = 'N'
```

```python
people_cols = ['project_cd', 'if_vrsn_id', 'plant_cd', 'res_cd', 'res_nm', 'res_seq', 'res_type_cd', 
        'res_ctgi_cd', 'res_grp_cd', 'res_capa_val', 'over_capa_val', 'capa_unit_cd', 
        'start_yymmdd', 'end_yymmdd', 'descr', 'inf_capa_yn', 'use_yn', 'del_yn']
```

```python
capa_people = pd.DataFrame({
    'res_cd': res_cd,
    'res_nm': res_cd,
    'res_capa_val': res_capa_val,
    'over_capa_val': over_capa_val
    }
)
```

```python
capa_people['project_cd'] = project_cd
capa_people['if_vrsn_id'] = if_vrsn_id
capa_people['plant_cd'] = plant_cd
capa_people['res_seq'] = res_seq
capa_people['res_type_cd'] = res_type_cd
capa_people['res_ctgi_cd'] = res_ctgi_cd
capa_people['res_grp_cd'] = res_grp_cd
capa_people['capa_unit_cd'] = capa_unit_cd
capa_people['start_yymmdd'] = start_yymmdd
capa_people['end_yymmdd'] = end_yymmdd
capa_people['descr'] = descr
capa_people['inf_capa_yn'] = inf_capa_yn
capa_people['use_yn'] = use_yn
capa_people['del_yn'] = del_yn
```

```python
capa_people = capa_people[people_cols]
capa_people
```

```python
capa_people.to_csv('res_people.csv', index=False)
```

### Machine Group <-> People Group Map

```python
project_cd = 'ENT_001'
plant_cd = 'K120'
res_grp_cd = ['2217'] * len(res_cd)
res_type_cd = ['PPL'] * len(res_cd)
use_yn = 'Y'
del_yn = 'N'
```

```python
res_map_cd = []
for res in res_org:
    res_map_cd.extend([res] * 20)
```

```python
map_cols = ['project_cd', 'plant_cd', 'res_map_cd', 'res_grp_cd','res_type_cd', 'res_cd', 'res_nm',
        'use_yn', 'del_yn']
```

```python
capa_map = pd.DataFrame({
    'res_map_cd': res_map_cd,
    'res_grp_cd': res_grp_cd,
    'res_type_cd': res_type_cd,
    'res_cd': res_cd,
    'res_nm': res_cd,
    'use_yn': use_yn,
    'del_yn': del_yn,
    }
)
```

```python
capa_map['project_cd'] = project_cd
capa_map['plant_cd'] = plant_cd
capa_map['use_yn'] = use_yn
capa_map['del_yn'] = del_yn
```

```python
capa_map = capa_map[map_cols]
capa_map = capa_map.sort_values(by=['plant_cd', 'res_map_cd', 'res_grp_cd', 'res_cd']).reset_index(drop=True)
```

```python
capa_map.to_csv('res_people_map.csv', index=False)
```
